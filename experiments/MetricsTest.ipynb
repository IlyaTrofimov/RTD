```python
barc = barcodes.calc_cross_barcodes(P, Q)
barcodes.plot_barcodes(barcodes.barc2array(barc), title = 'Cross-Barcodes')
```

```python
barc = barcodes.calc_embed_dist(P, Q, norm = 'quantile')
barcodes.plot_barcodes(barcodes.barc2array(barc), title = 'Embeddings\' Distance')
```
