```python
import os
import sys

os.environ['CUDA_VISIBLE_DEVICES'] = '2'
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

from sklearn.metrics.pairwise import pairwise_distances
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from cka import cka
from scipy import stats
from tqdm.notebook import tqdm
import seaborn as sns
%matplotlib inline

import numpy as np
import barcodes

def h1sum(barc):
    if 1 in barc:
        return sum([x[1] - x[0] for x in barc[1]])
    else:
        return 0.0
    
    
from sklearn.manifold import TSNE

def plot_tsne(x, title=''):
    tsne = TSNE(n_components=2, random_state=0)
    T = tsne.fit_transform(x)
    plt.scatter(T[:,0], T[:,1])
    plt.title(title)
    plt.show()
```

### Evaluation

```python
acc = (y == y_true).mean(1)
dis = (y != y[-1]).mean(1) / (1 - 0.5*acc - 0.5*acc[-1])

np.save('results_quantile/evolution_dis.npy', dis)
```

```python
sums = np.zeros((8, 8))
prods = np.zeros((8, 8))
ckas = np.ones((8, 8))

l = [0, 1, 3, 7, 13, 26, 51, 99]

for i in tqdm(range(len(l)), leave=True):
    for j in tqdm(range(i+1, len(l)), leave=False):
        x = X[l[i]]
        y = X[l[j]]
        barc = barcodes.calc_embed_dist(x, y, pdist_device = 'cuda:0', norm='quantile')
        s = h1sum(barc)

        barc = barcodes.calc_embed_dist(y, x, pdist_device = 'cuda:0', norm='quantile')
        s += h1sum(barc)

        sums[i,j] = s/2
        sums[j,i] = sums[i,j]

        ckas[i,j] = cka(x, y)
        ckas[j,i] = ckas[i,j]

np.save('results_quantile/vgg_evolution_h1sums.npy', sums)
np.save('results_quantile/vgg_evolution_ckas.npy', ckas)
```

# Ensembles

```python
embeddings = np.zeros((8, 10000, 64))

for path in os.listdir('embeddings/cifar10_embeddings/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_embeddings/{path}')
    if s == 'embedding':
        embeddings[int(n)-1] = x
    if s == 'embedding-fge':
        embeddings[int(n)-1 + 4] = x
    

predicts = np.zeros((8, 10000, 10))
for path in os.listdir('embeddings/cifar10_predicts/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_predicts/{path}')
    if s == 'predict':
        predicts[int(n)-1] = x
    if s == 'predict-fge':
        predicts[int(n)-1 + 4] = x
y_true = np.load('embeddings/cifar10_predicts/true_labels.npy')
        
        
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
y = predicts[:, idx].argmax(-1)
y_true = y_true[idx]
```

### Evaluation

```python
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0', norm='quantile')
        C[i, j] = h1sum(barc)
        
C = (C + C.T) / 2
np.save('results_quantile/ensemble_h1sum.npy', C)
```

```python
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        C[i, j] = cka(X[i], X[j])
        
np.save('results_quantile/ensemble_cka.npy', C)
```

```python
accs = (y == y_true.reshape(1, -1)).mean(1)
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        C[i, j] = (y[i] != y[j]).sum()
        
np.save('results_quantile/ensemble_dis.npy', C)
```

```python
from pathlib import Path

N_LAUNCHES = 5

save_dir = 'results_mean/noise'

np.random.seed(0)
seeds = np.random.randint(2**31, size=N_LAUNCHES+100, dtype=np.uint32)
seeds = np.random.choice(seeds, size=N_LAUNCHES, replace=False)
np.random.seed()
print(seeds)

for n_launch in range(1, N_LAUNCHES+1):
    np.random.seed(seeds[n_launch-1])
    idx = np.random.choice(range(10000), size=1000, replace=False)
    X = embeddings[:, idx]
    y = predicts[:, idx]
    y_true = Y_TRUE[idx]
    
    Path(f'{save_dir}/{n_launch}').mkdir(parents=True, exist_ok=True)
    
    accs = (y.argmax(-1) == y_true).mean(1) * 100
    np.save(f'{save_dir}/{n_launch}/noise_acc.npy', accs)
    np.save(f'{save_dir}/{n_launch}/noise_degrees', degrees)


    sums = []
    ckas = []
    for i in tqdm(range(1, len(X))):
        x = X[0]
        y = X[i]
        barc = barcodes.calc_embed_dist(x, y, pdist_device = 'cuda:0', norm='quantile')
        s = h1sum(barc)

        barc = barcodes.calc_embed_dist(y, x, pdist_device = 'cuda:0', norm='quantile')
        s += h1sum(barc)

        sums.append(s/2)
        
        s = cka(x, y)
        ckas.append(s)

    np.save(f'{save_dir}/{n_launch}/noise_h1sums.npy', sums)
    np.save(f'{save_dir}/{n_launch}/noise_ckas.npy', ckas)

    print(f'{n_launch} launch complete')
```

```python
strengths = np.load('results_quantile/saturation_strengths.npy')
accs = np.load('results_quantile/saturation_acc.npy')
sums = np.load('results_quantile/saturation_h1sums.npy')
ckas = np.load('results_quantile/saturation_ckas.npy')

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

ax[0, 0].plot(strengths[1:], sums[::-1], lw=2)
ax[0, 0].set_ylabel('h1sum')
               
ax[0, 1].plot(strengths[1:], ckas[::-1], lw=2)
ax[0, 1].set_ylabel('CKA')

ax[1, 0].plot(strengths, accs[::-1], lw=2)
ax[1, 0].set_ylabel('Accuracy')

plt.show()
```

# Transfer Learning

```python
dirs = [
    'embedding_100_100.npy',
    'embedding_100_10.npy',
    'embedding_10_100.npy',
    'embedding_10_10.npy']

embeddings = np.zeros((4, 10000, 64))
for n, path in enumerate(dirs):
    x = np.load(f'embeddings/cifar_transfer/{path}')
    embeddings[n] = x
    
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
```

```python
i = 0
j = 2

tsne = TSNE(n_components=2, random_state=0)

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

T = tsne.fit_transform(X[i])
ax[0, 0].scatter(T[:,0], T[:,1])
ax[0, 0].set_title(f'CIFAR-100 before')

T = tsne.fit_transform(X[j])
ax[0, 1].scatter(T[:,0], T[:,1])
ax[0, 1].set_title(f'CIFAR-100 after')


barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0', norm='quantile')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'before -> after, h1sum: {s:.2f}', ax=ax[1, 0], fig=fig)


barc = barcodes.calc_embed_dist(X[j], X[i], pdist_device = 'cuda:0', norm='quantile')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'after -> before, h1sum: {s:.2f}', ax=ax[1, 1], fig=fig)

ylim = max(ax[1,0].get_ylim()[0], ax[1,1].get_ylim()[0]), max(ax[1,0].get_ylim()[1], ax[1,1].get_ylim()[1])
ax[1, 0].set_ylim(*ylim)
ax[1, 1].set_ylim(*ylim)


plt.show()
```

```python
i = 1
j = 3

tsne = TSNE(n_components=2, random_state=0)

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

T = tsne.fit_transform(X[i])
ax[0, 0].scatter(T[:,0], T[:,1])
ax[0, 0].set_title(f'CIFAR-10 before')

T = tsne.fit_transform(X[j])
ax[0, 1].scatter(T[:,0], T[:,1])
ax[0, 1].set_title(f'CIFAR-10 after')


barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0', norm='quantile')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'before -> after, h1sum: {s:.2f}', ax=ax[1, 0], fig=fig)


barc = barcodes.calc_embed_dist(X[j], X[i], pdist_device = 'cuda:0', norm='quantile')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'after -> before, h1sum: {s:.2f}', ax=ax[1, 1], fig=fig)

ylim = max(ax[1,0].get_ylim()[0], ax[1,1].get_ylim()[0]), max(ax[1,0].get_ylim()[1], ax[1,1].get_ylim()[1])
ax[1, 0].set_ylim(*ylim)
ax[1, 1].set_ylim(*ylim)


plt.show()
```
