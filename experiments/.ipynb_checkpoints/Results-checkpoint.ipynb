```python
import os
import sys

os.environ['CUDA_VISIBLE_DEVICES'] = '2'

from sklearn.metrics.pairwise import pairwise_distances
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from cka import cka
from scipy import stats
from tqdm.notebook import tqdm
import seaborn as sns
%matplotlib inline

import numpy as np
import barcodes

def h1sum(barc):
    if 1 in barc:
        return sum([x[1] - x[0] for x in barc[1]])
    else:
        return 0.0
    
    
from sklearn.manifold import TSNE

def plot_tsne(x, title=''):
    tsne = TSNE(n_components=2, random_state=0)
    T = tsne.fit_transform(x)
    plt.scatter(T[:,0], T[:,1])
    plt.title(title)
    plt.show()
```

# Evolution

```python
embeddings = np.zeros((101, 10000, 64))
predicts = np.zeros((101, 10000, 10))

for path in os.listdir('embeddings/cifar10_evolution/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_evolution/{path}')
    if s == 'embedding':
        embeddings[int(n)] = x
    if s == 'predict':
        predicts[int(n)] = x
```

```python
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
y = predicts[:, idx]
```

### Evaluation

```python
sums = []
for i in tqdm(range(1, len(X)-1)):
    x = X[i]
    y = X[-1]
    barc = barcodes.calc_embed_dist(x, y, pdist_device = 'cuda:0')
    s = h1sum(barc)
    
    barc = barcodes.calc_embed_dist(y, x, pdist_device = 'cuda:0')
    s += h1sum(barc)
    
    sums.append(s/2)

#np.save('results/evolution_h1sums.npy', sums)
```

```python
ckas = []
for i in tqdm(range(1, len(X)-1)):
    x = X[i]
    y = X[-1]
    s = cka(x, y)
    ckas.append(s)
#np.save('results/evolution_ckas.npy', ckas)
```

### Plots

```python
for k in np.arange(0, 101, 10):
    plot_tsne(X[k], title=f'{k} epoch')
```

```python
sums = np.load('results/evolution_h1sums.npy')
ckas = np.load('results/evolution_ckas.npy')

fig, ax = plt.subplots(1, 2, figsize=(16, 5))
ax[0].plot(sums, lw=2)
ax[0].set_title('h1sum')

ax[1].plot(ckas, lw=2)
ax[1].set_title('CKA')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(-(sums - sums.mean(0)) / sums.std(), lw=2, label='h1sum')
plt.plot((ckas - ckas.mean(0)) / ckas.std(), lw=2, label='CKA')
plt.legend()
plt.show()
```

# Ensembles

```python
embeddings = np.zeros((8, 10000, 64))

for path in os.listdir('embeddings/cifar10_embeddings/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_embeddings/{path}')
    if s == 'embedding':
        embeddings[int(n)-1] = x
    if s == 'embedding-fge':
        embeddings[int(n)-1 + 4] = x
    

predicts = np.zeros((8, 10000, 10))
for path in os.listdir('embeddings/cifar10_predicts/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_predicts/{path}')
    if s == 'predict':
        predicts[int(n)-1] = x
    if s == 'predict-fge':
        predicts[int(n)-1 + 4] = x
y_true = np.load('embeddings/cifar10_predicts/true_labels.npy')
        
        
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
y = predicts[:, idx].argmax(-1)
y_true = y_true[idx]
```

### Evaluation

```python
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0')
        C[i, j] = h1sum(barc)
        
C = (C + C.T) / 2
# np.save('results/ensemble_h1sum.npy', C)
```

```python
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        C[i, j] = cka(X[i], X[j])
        
#np.save('results/ensemble_cka.npy', C)
```

```python
accs = (y == y_true.reshape(1, -1)).mean(1)
C = np.zeros((8, 8))
for i in tqdm(range(8)):
    for j in range(8):
        if i == j: continue
        C[i, j] = (y[i] != y[j]).mean() / (1 - (accs[i] + accs[j]) * 0.5)
        
#np.save('results/ensemble_dis.npy', C)
```

### Plots

```python
C_h1 = np.load('results/ensemble_h1sum.npy')
C_cka = np.load('results/ensemble_cka.npy')
C_dis = np.load('results/ensemble_dis.npy')


fig, ax = plt.subplots(2, 2, figsize=(16, 11))
sns.heatmap(C_h1, annot=True, ax=ax[0, 0], cmap='Blues', mask=np.eye(8, 8, dtype='bool'))
ax[0, 0].set_title('h1sum')

sns.heatmap(C_cka, annot=True, ax=ax[0, 1], cmap='Blues', mask=np.eye(8, 8, dtype='bool'))
ax[0, 1].set_title('CKA')

sns.heatmap(C_dis, annot=True, ax=ax[1, 0], cmap='Blues', mask=np.eye(8, 8, dtype='bool'))
ax[1, 0].set_title('Disagreement')

plt.show()
```

# Noise

```python
embeddings = np.zeros((61, 10000, 64))
predicts = np.zeros((61, 10000, 10))

for path in os.listdir('embeddings/cifar10_noise/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_noise/{path}')
    if s == 'embedding':
        embeddings[int(n)] = x
    if s == 'predict':
        predicts[int(n)] = x
y_true = np.load('embeddings/cifar10_predicts/true_labels.npy')

degrees = np.linspace(0, 1, 61)

np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
y = predicts[:, idx]
y_true = y_true[idx]
```

```python
accs = (y.argmax(-1) == y_true).mean(1) * 100
np.save('results/noise_acc.npy', accs)
np.save('results/noise_degrees', degrees)


sums = []
for i in tqdm(range(1, len(X))):
    x = X[0]
    y = X[i]
    barc = barcodes.calc_embed_dist(x, y, pdist_device = 'cuda:0')
    s = h1sum(barc)
    
    barc = barcodes.calc_embed_dist(y, x, pdist_device = 'cuda:0')
    s += h1sum(barc)
    
    sums.append(s/2)

np.save('results/noise_h1sums.npy', sums)


ckas = []
for i in tqdm(range(1, len(X))):
    x = X[0]
    y = X[i]
    s = cka(x, y)
    ckas.append(s)
np.save('results/noise_ckas.npy', ckas)
```

```python
degrees = np.load('results/noise_degrees.npy')
accs = np.load('results/noise_acc.npy')
sums = np.load('results/noise_h1sums.npy')
ckas = np.load('results/noise_ckas.npy')

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

ax[0, 0].plot(degrees[1:], sums, lw=2)
ax[0, 0].set_ylabel('h1sum')
               
ax[0, 1].plot(degrees[1:], ckas, lw=2)
ax[0, 1].set_ylabel('CKA')

ax[1, 0].plot(degrees, accs, lw=2)
ax[1, 0].set_ylabel('Accuracy')

plt.show()
```

# Transfer Learning

```python
dirs = [
    'embedding_100_100.npy',
    'embedding_100_10.npy',
    'embedding_10_100.npy',
    'embedding_10_10.npy']

embeddings = np.zeros((4, 10000, 64))
for n, path in enumerate(dirs):
    x = np.load(f'embeddings/cifar_transfer/{path}')
    embeddings[n] = x
    
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
```

```python
i = 0
j = 2

tsne = TSNE(n_components=2, random_state=0)

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

T = tsne.fit_transform(X[i])
ax[0, 0].scatter(T[:,0], T[:,1])
ax[0, 0].set_title(f'CIFAR-100 before')

T = tsne.fit_transform(X[j])
ax[0, 1].scatter(T[:,0], T[:,1])
ax[0, 1].set_title(f'CIFAR-100 after')


barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'before -> after, h1sum: {s:.2f}', ax=ax[1, 0], fig=fig)


barc = barcodes.calc_embed_dist(X[j], X[i], pdist_device = 'cuda:0')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'after -> before, h1sum: {s:.2f}', ax=ax[1, 1], fig=fig)

ylim = max(ax[1,0].get_ylim()[0], ax[1,1].get_ylim()[0]), max(ax[1,0].get_ylim()[1], ax[1,1].get_ylim()[1])
ax[1, 0].set_ylim(*ylim)
ax[1, 1].set_ylim(*ylim)


plt.show()
```

```python
i = 1
j = 3

tsne = TSNE(n_components=2, random_state=0)

fig, ax = plt.subplots(2, 2, figsize=(16, 11))

T = tsne.fit_transform(X[i])
ax[0, 0].scatter(T[:,0], T[:,1])
ax[0, 0].set_title(f'CIFAR-10 before')

T = tsne.fit_transform(X[j])
ax[0, 1].scatter(T[:,0], T[:,1])
ax[0, 1].set_title(f'CIFAR-10 after')


barc = barcodes.calc_embed_dist(X[i], X[j], pdist_device = 'cuda:0')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'before -> after, h1sum: {s:.2f}', ax=ax[1, 0], fig=fig)


barc = barcodes.calc_embed_dist(X[j], X[i], pdist_device = 'cuda:0')
s = h1sum(barc)
barc = barcodes.barc2array(barc)
barcodes.plot_barcodes(barc, title=f'after -> before, h1sum: {s:.2f}', ax=ax[1, 1], fig=fig)

ylim = max(ax[1,0].get_ylim()[0], ax[1,1].get_ylim()[0]), max(ax[1,0].get_ylim()[1], ax[1,1].get_ylim()[1])
ax[1, 0].set_ylim(*ylim)
ax[1, 1].set_ylim(*ylim)


plt.show()
```

# Minimize CKA

```python
embeddings = np.zeros((3, 10000, 64))
predicts = np.zeros((3, 10000, 10))

for path in os.listdir('embeddings/cifar10_minimize_cka/'):
    s, n = path.split('.')[0].split('_')
    x = np.load(f'embeddings/cifar10_minimize_cka/{path}')
    if s == 'embedding':
        embeddings[int(n)-1] = x
    if s == 'predict':
        predicts[int(n)-1] = x
        
y_true = np.load('embeddings/cifar10_predicts/true_labels.npy')
        
np.random.seed(0)
idx = np.random.choice(range(10000), size=1000, replace=False)
X = embeddings[:, idx]
y = predicts[:, idx].argmax(-1)
y_true = y_true[idx]
```
